```python
import pandas as pd
import json
from google.colab import drive
drive.mount('/content/drive')
```

```python
f = open('/content/drive/My Drive/Colab Notebooks/orders_ETL.json') # f is nothing but a file object
data = json.load(f)
data
```

```python
len(data) # len of data is 12 which means we have 12 order_id's
```

```python
# we need to male 3 tables
# 1. orders_table, customer_table & product_table below mentioned are the column required in our table as per data
# ---> order_id, order_date, total_amount, quantity, customer_id, product_id
# ---> customer id, name, email, address
# ---> products_id, name, category, price
```

```python
#order table
order_id = []
order_date = []
total_amount = []
quantity = []
customer_id = []
product_id = []


for oder in data:
  for product in oder['products']:
    order_id.append(oder['order_id'])
    order_date.append(oder['order_date'])
    total_amount.append(oder['total_amount'])
    customer_id.append(oder['customer']['customer_id']) # dic inside dic
    product_id.append(product['product_id'])
    quantity.append(product['quantity'])

df_orders = pd.DataFrame({

    "order_id": order_id,
    "order_date": order_date,
    "total_amount": total_amount,
    "customer_id": customer_id,
    "product_id": product_id,
    "quantity": quantity

})

df_orders
```

```python
# product table

product_id = []
product_name = []
product_category = []
product_price = []

for orders in data:
  for product in orders['products']:
    product_id.append(product['product_id'])
    product_name.append(product['name'])
    product_category.append(product['category'])
    product_price.append(product['price'])


df_product = pd.DataFrame({
    "product_id": product_id,
    "product_name": product_name,
    "product_category": product_category,
    "product_price": product_price
})

# we have to remove the duplicates in product_id

df_product.drop_duplicates(subset = 'product_id', keep = 'last', inplace = True)
df_product


```

```python
# customer_table

customer_id = []
customer_name = []
customer_email = []
customer_address = []

for order in data:
  customer_id.append(order['customer']['customer_id'])
  customer_name.append(order['customer']['name'])
  customer_email.append(order['customer']['email'])
  customer_address.append(order['customer']['address'])


df_customer = pd.DataFrame({

    "customer_id": customer_id,
    "customer_name": customer_name,
    "customer_email": customer_email,
    "customer_address": customer_address
})


# here we can see duplications, so when we remove the duplicates we need to keep the last record

df_customer.drop_duplicates(subset = 'customer_id', keep = 'last', inplace = True)
df_customer
```

```python
from sqlalchemy import create_engine, text

engine = create_engine('mysql+pymysql://root:password@localhost:3306/orders')

df_orders.to_sql(name='orders_table', con = engine, if_exists = 'replace', index = False)
df_customer.to_sql(name='cutomers_table', con = engine, if_exists = 'replace', index = False)
df_product.to_sql(name='products_table', con = engine, if_exists = 'replace', index = False)
```
